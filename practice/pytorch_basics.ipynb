# Introduction to PyTorch

```
import torch
```

### torch.empty()

```
# Returns a tensor filled with uninitialized data. The shape of the tensor is defined by the variable argument size
x = torch.empty(5, 3)
print(x)
print(x.size())
print(type(x))
print(x[0][0])
```

```
print(x + 1)
```

### torch.rand()

```
# Returns a tensor filled with random numbers from a uniform distribution on the interval [0,1)
x = torch.rand(5, 3)
print(x)
print(x.size())
print(type(x))
print(x[0][0])
```

```
print(x + 1)
print(x - 1)
```

```
print(torch.add(x,1))
print(torch.sub(x,1))
```

### torch.Tensor()

```
x = torch.rand(5, 3)
result = torch.Tensor(5, 3)
print(x)
print(result)
```

```
torch.add(x, 1, out=result)
print(result)
torch.sub(x, 1, out=result)
print(result)
torch.mul(x, 2, out=result)
print(result)
torch.div(x, 2, out=result)
print(result)
```

```
x = torch.rand(5, 3)
print(x.add(1))
print(x)
```

```
print(x.add_(1)) # mutates the tensor in place
print(x)
```

```
print(x.add_(1)) # mutates the tensor in place
print(x.sub_(1))
print(x.mul_(1))
print(x.div_(1))
```

```
print(x[1,:])
```

### torch.ones() & torch.zeros()

```
ones = torch.ones(5, 3)
zeros = torch.zeros(5, 3)
print(ones)
print(zeros)
```
